```python
import os
import requests
from bs4 import BeautifulSoup

# URL dari halaman pencarian arXiv
url = 'https://arxiv.org/list/econ.GN/2404?skip=0&show=100'

# Lokasi folder untuk menyimpan file PDF
folder_location = r'./pdfs'
if not os.path.exists(folder_location):
    os.mkdir(folder_location)

def extract_and_save_pdf(url, folder_location):
    # Mengirim request ke URL dan mendapatkan response
    response = requests.get(url)
    
    # Parsing konten HTML menggunakan BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Mencari semua link PDF yang mungkin
    pdf_links = soup.find_all('a', href=True)
    pdf_links = [link for link in pdf_links if '/pdf/' in link['href']]  # Memfilter untuk memastikan itu link PDF

    for link in pdf_links:
        # Mendapatkan URL lengkap dari PDF
        full_url = 'https://arxiv.org' + link['href'] if not link['href'].startswith('http') else link['href']
        
        # Menentukan nama file
        filename = link['href'].split('/')[-1] + '.pdf'  # Menambahkan '.pdf' untuk memastikan ekstensi file
        
        # Mengunduh PDF dan menyimpannya sebagai file lokal
        with open(os.path.join(folder_location, filename), 'wb') as f:
            f.write(requests.get(full_url).content)
        
        print(f'Downloaded {filename}')

# Panggil fungsi untuk ekstraksi dan penyimpanan PDF
extract_and_save_pdf(url, folder_location)
```

```python
import shutil

# Tentukan path folder yang ingin dikompres dan nama file zip output
folder_path = 'pdfs'
output_filename = 'ArxivPaper_General Economics'

# Membuat file zip
shutil.make_archive(output_filename, 'zip', folder_path)
```
